Define x and y

```python
x = df.drop(['Survived'],axis=1)
y = df['Survived']
```

Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
cat = []
con = []

for i in x.columns:
    if x[i].dtypes == 'object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
num_pipe = Pipeline(steps = [('impute',SimpleImputer(strategy = 'median')),('scaler',StandardScaler())])
cat_pipe = Pipeline(steps = [('impute',SimpleImputer(strategy = 'most_frequent')),('encode',OneHotEncoder())])
```

```python
pre = ColumnTransformer([('num_pipe',num_pipe,con),('cat_pipe',cat_pipe,cat)])
```

```python
x1 = pre.fit_transform(x)
x1
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
x2 = pd.DataFrame(x1,columns=cols)
x2
```

Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x2,y,test_size = 0.2,random_state = 23)
```

Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
la = LogisticRegression()
kn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
ab = AdaBoostClassifier()
sv = SVC()
```

```python
list1 = [la,kn,dt,rf,ab,sv]
```

```python
for i in list1:
    model = i.fit(x_train,y_train)

    y_pred = model.predict(x_test)

    from sklearn.metrics import f1_score,accuracy_score
    acc = accuracy_score(y_pred,y_test)
    f = f1_score(y_pred,y_test)

    print('*'*50)
    print(i)
    print('Accuracy:',acc)
    print('f1_score:',f)
    print('*'*50)
```
